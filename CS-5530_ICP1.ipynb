Importing Libraries

```python
import pandas as pd
import numpy as np
```

Save the dataset to a variable and verify download by printing df

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df
```

Get all columns that have numeric data

```python
num_df = df.select_dtypes(include=np.number)
num_df
```

For loop to select all column names in the numeric dataframe, then get all of those values from the full dataframe. <br> Basically just gets the number of NaN
in numeric cols from the original df. We will use this to test if it worked later.

```python
df[[col for col in num_df.columns]].isna().sum()
```

get the mean for each column of the numeric df. We will use this to compare the means later.

```python
df[[col for col in num_df.columns]].mean()
```

Go through each column in the numerical df and replace each NaN with its corresponding mean value. <br> <br> **This is where I replace all NaN with the mean for that column.**

```python
for col in num_df.columns: # Loops through each column in the numerical df
  df[col] = df[col].fillna(df[col].mean()) # Fills the NaN of each column with the mean of that column
```

Now we can test if the values were overwritten in the original df. If they are all 0, then it worked.

```python
df[[col for col in num_df.columns]].isna().sum()
```

Since all of the values in the table are 0, that means there are no NaN values left. All we need to do now is to check if the mean of the column changed. If the mean is still the same then that shows only the mean values were added to the original df.

```python
df[[col for col in num_df.columns]].mean()
```

Since the means are the same, and it shows there are no more NaN values in the original df, it means we have successfully replaced all of the NaN values in the numerical columns with the mean for that column.
