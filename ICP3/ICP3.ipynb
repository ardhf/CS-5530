```python
import pandas as pd
```

## Loading the Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/JuliaData/CSV.jl/refs/heads/main/test/testfiles/Sacramentorealestatetransactions.csv')
df.head()
```

## Finding the Number of Columns in the dataset

```python
num_cols = len(df.columns)
print(f'This dataset has {num_cols} columns')
```

## Finding the Mean of the House Prices

```python
mean_house_price = df.price.mean()
print(f'The mean price of houses in this dataset is ${mean_house_price:,.2f}')
```

## Finding the Median of the House Prices

```python
median_house_price = df.price.median()
print(f'The median price of houses in this dataset is ${median_house_price:,.2f}')
```

## Finding the Different Types of Property in the Dataset

```python
different_property_types = df.type.unique()
formatted_prop_types = ', '.join(map(str, different_property_types))
print(f'The different types of properties in the dataset are: {formatted_prop_types}')
```

## Finding the Max Price of a Condo in the Dataset

```python
max_condo_price = df[(df.type == 'Condo')].price.max()
print(f'The max price of a condo is ${max_condo_price:,.2f}')
```

## Find the Mean of Each Type of Property in the Dataset

```python
mean_prop_type_prices = df[['price', 'type']].groupby("type").mean()
mean_prop_type_prices
```

## Finding the Count of Different Property Types in the Dataset

```python
prop_types_count = df.type.value_counts()
prop_types_count
```
